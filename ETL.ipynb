```python
# Re-import necessary libraries in the current environment
import os
import pandas as pd

# path to the "data" directory
data_folder_path = './data'

# Path to the Excel file
excel_file_path = os.path.join(data_folder_path, 'CM_Data_Explorer.xlsx')
sheet = '2.3 EV'

# Load the data into a pandas DataFrame for stated policies
df_stated_policies = pd.read_excel(excel_file_path, sheet_name=sheet, usecols="A:I", skiprows=81, nrows=11)

# Define the header row for stated policies
header_stated_policies = pd.read_excel(excel_file_path, sheet_name=sheet, usecols="B:I", skiprows=4, nrows=1).columns

# Set the header of the DataFrame for stated policies
df_stated_policies.columns = ['Mineral'] + list(header_stated_policies)

# Set the index of the DataFrame for stated policies
df_stated_policies.set_index('Mineral', inplace=True)

# Drop columns that contain NaN values for stated policies
df_stated_policies_cleaned = df_stated_policies.dropna(axis=1)

# Add a column to indicate the scenario for stated policies
df_stated_policies_cleaned['Scenario'] = 'Stated policies'
```

```python
import pandas as pd

# path to the "data" directory
data_folder_path = './data'

# Path to the Excel file
excel_file_path = os.path.join(data_folder_path, 'CM_Data_Explorer.xlsx')

# Load the 'Announced Pledge' scenario data
# Since the data starts from column K, which corresponds to 'usecols' 10 to 15 (0-indexed)
df_announced_pledge = pd.read_excel(
    excel_file_path,
    sheet_name='2.3 EV',
    usecols="K:P",  # Select columns K to P
    skiprows=82,  # Skip rows to start from row 83 (0-indexed)
    nrows=11,  # Read 12 rows starting from row 83 (1-indexed)
    header=None  # Do not treat the first row as the header
)

# Define the header row for stated policies
header_announced_pledge = pd.read_excel(excel_file_path, sheet_name=sheet, usecols="D:I", skiprows=4, nrows=1).columns

# Add the header to the DataFrame
df_announced_pledge.columns = header_announced_pledge

# Now, let's retrieve the index from the 'Stated Policies' DataFrame
index_stated_policies = df_stated_policies.index.tolist()

# Insert the index column to 'Announced Pledge' DataFrame
df_announced_pledge.insert(0, 'Mineral', index_stated_policies)

# Set the new column as the index
df_announced_pledge.set_index('Mineral', inplace=True)

df_announced_pledge

df_announced_pledge['Scenario'] = 'Announced pledges'
```

```python
df_announced_pledge
```

```python
# Load the 'Announced Pledge' scenario data
# Since the data starts from column K, which corresponds to 'usecols' 10 to 15 (0-indexed)
df_net_zero = pd.read_excel(
    excel_file_path,
    sheet_name='2.3 EV',
    usecols="R:W",  # Select columns K to P
    skiprows=82,  # Skip rows to start from row 83 (0-indexed)
    nrows=11,  # Read 12 rows starting from row 83 (1-indexed)
    header=None  # Do not treat the first row as the header
)

# Define the header row for stated policies
header_announced_pledge = pd.read_excel(excel_file_path, sheet_name=sheet, usecols="D:I", skiprows=4, nrows=1).columns

# Add the header to the DataFrame
df_net_zero.columns = header_announced_pledge

# Now, let's retrieve the index from the 'Stated Policies' DataFrame
index_stated_policies = df_stated_policies.index.tolist()

# Insert the index column to 'Announced Pledge' DataFrame
df_net_zero.insert(0, 'Mineral', index_stated_policies)

# Set the new column as the index
df_net_zero.set_index('Mineral', inplace=True)

df_net_zero

df_net_zero['Scenario'] = 'Net zero 2050'
```

```python
df_net_zero
```

```python
# Convert df_stated_policies_cleaned to long format
df_stated_long = pd.melt(df_stated_policies_cleaned.reset_index(), id_vars=['Mineral', 'Scenario'], var_name='Year', value_name='Demand')
df_stated_long['Year'] = df_stated_long['Year'].astype(int)  # Ensure 'Year' is integer

# Convert df_announced_pledge to long format
df_announced_long = pd.melt(df_announced_pledge.reset_index(), id_vars=['Mineral', 'Scenario'], var_name='Year', value_name='Demand')
df_announced_long['Year'] = df_announced_long['Year'].astype(int)

# Assuming df_net_zero is prepared similarly
df_net_zero_long = pd.melt(df_net_zero.reset_index(), id_vars=['Mineral', 'Scenario'], var_name='Year', value_name='Demand')
df_net_zero_long['Year'] = df_net_zero_long['Year'].astype(int)

# Merge all three DataFrames
merged_df = pd.concat([df_stated_long, df_announced_long, df_net_zero_long])

# If you want to set 'Mineral' and 'Year' as a MultiIndex
merged_df.set_index(['Mineral', 'Year'], inplace=True)

# Show the merged DataFrame
print(merged_df.head())
```

```python
merged_df
```

```python
# Assuming merged_df is your DataFrame
merged_df.columns = [col.lower() for col in merged_df.columns]

# Or equivalently
merged_df.columns = merged_df.columns.str.lower()

# Show the updated DataFrame columns to confirm
print(merged_df.columns)
```

```python
merged_df_reset = merged_df.reset_index()

# Now let's convert all column names to lowercase
merged_df_reset.columns = merged_df_reset.columns.str.lower()

merged_df_reset
```

```python
# Export the merged DataFrame to a CSV file
merged_df_reset.to_csv('minerals_demand_long.csv', index=False)
```

```python
# Worldwide Reserves
excel_file_path = os.path.join(data_folder_path, 'cobalt-reserves.xlsx')
sheet = 'Data'

cobalt_reserves = pd.read_excel(excel_file_path, sheet_name=sheet, usecols="B:C", skiprows = 4, nrows=23)
```

```python
# Worldwide Reserves
excel_file_path = os.path.join(data_folder_path, 'lithium-reserves.xlsx')
sheet = 'Data'

lithium_reserves = pd.read_excel(excel_file_path, sheet_name=sheet, usecols="B:C", skiprows = 4, nrows=23)
```

```python
cobalt_reserves
cobalt_reserves.columns = ['year', 'reserve_kt']
cobalt_reserves['reserve_kt'] = cobalt_reserves['reserve_kt'] * 1000
cobalt_reserves['mineral']='Cobalt'
```

```python
lithium_reserves
lithium_reserves.columns = ['year', 'reserve_kt']
lithium_reserves['reserve_kt'] = lithium_reserves['reserve_kt'] * 1000
lithium_reserves['mineral']='Lithium'
```

```python
cobalt_reserves
```

```python
lithium_reserves
```

```python
nickel_reserves
```

```python
# Concatenate the dataframes
merged_reserves = pd.concat([cobalt_reserves, lithium_reserves, nickel_reserves])

# Resetting the index after concatenation for a clean DataFrame
merged_reserves.reset_index(drop=False, inplace=True)

merged_reserves
```

```python
# Sort the DataFrame to ensure it's in the correct order for calculations
merged_reserves.sort_values(by=['mineral', 'year'], inplace=True)

merged_reserves['new_finds_pct'] = merged_reserves.groupby('mineral')['reserve_kt'].pct_change().fillna(0) * 100

# Round the 'new_finds_pct' column to one decimal place
merged_reserves['new_finds_pct'] = merged_reserves['new_finds_pct'].round(1)
```
