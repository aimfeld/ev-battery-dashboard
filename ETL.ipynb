```python
import pandas as pd

# path to the "data" directory
data_folder_path = './data'

# Path to the Excel file
excel_file_path = os.path.join(data_folder_path, 'CM_Data_Explorer.xlsx')

# Load the 'Announced Pledge' scenario data
# Since the data starts from column K, which corresponds to 'usecols' 10 to 15 (0-indexed)
df_announced_pledge = pd.read_excel(
    excel_file_path,
    sheet_name='2.3 EV',
    usecols="K:P",  # Select columns K to P
    skiprows=82,  # Skip rows to start from row 83 (0-indexed)
    nrows=11,  # Read 12 rows starting from row 83 (1-indexed)
    header=None  # Do not treat the first row as the header
)

# Define the header row for stated policies
header_announced_pledge = pd.read_excel(excel_file_path, sheet_name=sheet, usecols="D:I", skiprows=4, nrows=1).columns

# Add the header to the DataFrame
df_announced_pledge.columns = header_announced_pledge

# Now, let's retrieve the index from the 'Stated Policies' DataFrame
index_stated_policies = df_stated_policies.index.tolist()

# Insert the index column to 'Announced Pledge' DataFrame
df_announced_pledge.insert(0, 'Mineral', index_stated_policies)

# Set the new column as the index
df_announced_pledge.set_index('Mineral', inplace=True)

df_announced_pledge

df_announced_pledge['Scenario'] = 'Announced pledges'
```

```python
# Before using concat, set the 'year' column as the index for each dataframe
nickel_reserves.set_index('year', inplace=True)
cobalt_reserves.set_index('year', inplace=True)
lithium_reserves.set_index('year', inplace=True)

# Now we can concatenate them along the columns axis (axis=1)
merged_df_res = pd.concat([nickel_reserves, cobalt_reserves, lithium_reserves], axis=1)

merged_df_res
```

```python
merged_df_res_cleaned = merged_df_res.dropna()
```

```python
merged_df_res_cleaned
```

```python
# Assuming merged_df_res_cleaned is your DataFrame
import numpy as np 

# Add new columns initialized with NaN for the new finds rate
for mineral in ['nickel_reserve_kt', 'cobalt_reserve_kt', 'lithium_reserve_kt']:
    merged_df_res_cleaned[f'new_finds_{mineral}'] = np.nan

# Calculate the new finds rate for each mineral for each year after 2010
for year in merged_df_res_cleaned.index[1:]:  # Skip the first year (2010)
    for mineral in ['nickel_reserve_kt', 'cobalt_reserve_kt', 'lithium_reserve_kt']:
        # Calculate the new finds rate
        new_finds = merged_df_res_cleaned.loc[year, mineral] - merged_df_res_cleaned.loc[year - 1, mineral]
        # Assign the new finds rate to the corresponding year and mineral
        merged_df_res_cleaned.loc[year, f'new_finds_{mineral}'] = new_finds

# Set the first year's new finds to 0 as per the requirement
for mineral in ['nickel_reserve_kt', 'cobalt_reserve_kt', 'lithium_reserve_kt']:
    merged_df_res_cleaned.loc[2010, f'new_finds_{mineral}'] = 0

merged_df_res_cleaned
```
